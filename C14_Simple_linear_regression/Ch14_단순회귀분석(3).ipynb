# 최대우도추정법
---

## 개념
- 최대우도추정 : 모수가 미지의 θ 인 확률분포에서 뽑은 표본 x 들을 바탕으로 θ 를 추정하는 기법
- 우도 : 이미 주어진 표본 x 들에 비추어 봤을 때, 모집단의 모수 θ 에 대한 추정이 그럴듯한 정도
  - 관측치를 고정시키고 그런 관측치가 나오게 하는 가장 그럴 듯한 모수값을 추정
  - 우도 L(θ|x)는 θ가 전제되었을 때 표본 x가 등장할 확률인 p(x|θ)에 비례
- 사용 이유
  - 최대우도추정 기법으로 추정한 모수는 일치성과 효율성을 가짐
  - 일치성(consistency) : 추정에 사용하는 표본의 크기가 커질 수록 진짜 모수값에 수렴하는 특성
  - 효율성(efficiency) : 일치성 등에서 같은 추정량 가운데서도 분산이 작은 특성
  - 머신러닝에서 모수 추정시 최대우도추정 기법을 많이 사용 -> 표본에 큰 영향을 받기 때문에 이를 보완하는 기법들이 존재 





## 최소자승법을 사용한 이유 -> 최대우도추정법
- 단순 선형회귀분석에서 최소 자승법을 이용해서 x 와 y 의 관계를 설명
  - 즉, 회귀식 추정 y 와 실제 관측 y 의 오차를 최소화 할 수 있는 회귀식을 적합
- 최대우도법을 이용해 회귀식 추정이 가능
- 관측치 (x_i, y_i) 에 대해 우도를 구함 -> 전체 데이터에 대한 우도는 각 데이터에 대한 우도를 모두 곱한 값
- 오류의 제곱 값을 최소화하는 알파, 베타가 계산되는 지점이 우도가 최대화되는 지점
- 즉, 오류의 제곱 값을 최소화하는 것은 관측된 데이터가 발생할 우도를 최대화하는 것과 동일
