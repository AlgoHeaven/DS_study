# Gradient 적용시키기
---

- 경사도 벡터를 0으로 만들기

```python
import random
import math

def step(v, direction, step_size):
  return [v_i + step_size * direction_i for v_i, direction_i in zip(v, direction)]

def sum_of_squares_gradient(v):
  return [2 * v_i for v_i in v]

def distance(next_v, v):
  return math.sqrt(sum([(next_v_i - v_i)**2 for next_v_i, v_i in zip(next_v, v)])) 

# 임의의 시작점 선택
v = [random.randint(-10,10) for _ in range(3)] # 경사도를 나타내는 벡터

tolerance = 0.0000001
count = 0

print(v)
while True: # 경사도를 0으로 만드는 과정
  gradient = sum_of_squares_gradient(v)
  next_v = step(v, gradient, -0.01)
  if distance(next_v, v) < tolerance:
    break
  v = next_v
  count+=1
  if count%20==0:
    print(next_v)
```

# 적절한 이동 거리 구하기
---

- step size에 따른 error catch

```python
step_size = [100, 10, 1, 0.1, 0.01, 0.001, 0.0001, 0.00001]

for s in step_size[:3]:
  v = [8,1,10]
  try:
    while True:
      gradient = sum_of_squares_gradient(v)
      next_v = step(v, gradient, s)
      if distance(next_v, v) < tolerance:
        break
      v = next_v
      count+=1
      if count%10==0:
        print(next_v)
  except:
    print(float('inf'))
```
